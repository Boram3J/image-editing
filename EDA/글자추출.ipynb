```python
# 전처리 함수
import matplotlib.pyplot as plt

def preprocessing(org_image):

    # SRGAN 사용하여 해상도를 높이는 코드: SRGAN_zzinmak.py 
    # 근데 이미지가 안 넣어짐..왜지...-> 크기가 너무 커서!!!. 넘겨줄 때 크기가 좀 작은 거 넘겨줘야 함.
    #srgan = SRGAN_zzinmak.apply_srgan(org_image)
    # srgan 안 쓰는 게 더 나음. 근데 왤까??? 화질이 srgan_zzinmak에서 돌리면 이미지가 더 뚜렷해지는데..

    # srgan을 적용시키는 법을 고민해보자.


    # 그레이 스케일 이미지
    gray_img = cv2.cvtColor(org_image, cv2.COLOR_BGR2GRAY)

    # grayscale -> binary
    binary_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    # plt_imshow('binary', binary_img)

    # noise remove
    kernel_size = 3 # 커널 사이즈를 홀수로 해야 함!!!!
    noNoise = cv2.medianBlur(gray_img, ksize=kernel_size)
    #plt_imshow('blur', noNoise)



    #plt.imshow(noNoise)
    return noNoise
```

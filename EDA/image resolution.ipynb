```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# 이미지 로드
img = cv2.imread('.\\imageData\\michaeljackson.jpg')
```

```python
# b,g,r to r,g,b
def BGR2RGB(image):
    b,g,r = cv2.split(image)
    rgb_image = cv2.merge([r,g,b])
    return rgb_image
```

```python
# 이미지 추론
def upsample(path,model_name,set_model_param,image, xN):
    sr = cv2.dnn_superres.DnnSuperResImpl_create()
    sr.readModel(os.path.join(path,model_name))
    sr.setModel(set_model_param, xN)
    result = sr.upsample(image)
    return result
```

```python
# 결과 이미지 비교
def comparison_image(origin_img, result_img, xN): 
    # Resized image
    opencv_resized_img = cv2.resize(origin_img,dsize=None,fx=xN,fy=xN)
    
    cv2.imwrite(".\\result\\result_img.jpg", result_img)
    
    plt.figure(figsize=(12,8))

    plt.subplot(1,3,1)
    # Original image 원본사진
    plt.imshow(BGR2RGB(origin_img))
    plt.title('Original Image')
    plt.subplot(1,3,2)
    # SR upscaled 모델 통해 upscale
    plt.imshow(BGR2RGB(result_img))
    plt.title('Upscaled_Image_via_Model')
    plt.subplot(1,3,3)
    # OpenCV upscaled  
    plt.imshow(BGR2RGB(opencv_resized_img))
    plt.title('Upscaled_Image_via_Opencv_resize')
    plt.show()
```
